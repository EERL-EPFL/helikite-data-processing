```python
from helikite.classes.data_processing_level2 import DataProcessorLevel2
from helikite.config import load_config
from helikite.constants import constants

import pandas as pd
import glob
import os

config = load_config(constants.CONFIG_FILE)

# Define the folder containing your CSV files
folder = constants.OUTPUTS_FOLDER / "Processing" / "Level2"

# Find all CSV files in the folder
csv_files = glob.glob(os.path.join(folder, "*.csv"))

df_list = []

for i, file in enumerate(csv_files):
    # Read all files the same way; we'll parse datetime afterward
    df = DataProcessorLevel2.read_data(file)
    df_list.append(df)

# Combine all files
combined_df = pd.concat(df_list)
combined_df.sort_index(inplace=True)

# Save as a single comma-separated text file
output_path = folder / "ORACLES_64Helikite_2024.txt"
combined_df.to_csv(output_path, index=True)
```
